0) Sea A = {a, b, c, d, e, f} y R una relación sobre A así: R = {(a, a), (b, c), (c, b), (c, e)}. Represente como digrafo GR en networkx. Usando el algoritmo transitive_closure de networkx calcule el grafo de la cerradura transitiva de GR. Nota: El algoritmo requiere un digrafo que se crea con la clase Digraph en vez de Graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

a = 'a'
b = 'b'
c = 'c'
d = 'd'
e = 'e'
f = 'f'

A = [a, b, c, d, e, f]
R = [(a,a),(b,c),(c,b),(c,e)]

digraph = nx.DiGraph(R)
nx.draw(digraph, with_labels=True)
plt.show()


transitiveClosure = nx.transitive_closure(digraph, reflexive=False)
print(transitiveClosure.edges())


nx.draw(transitiveClosure, with_labels=True)
plt.show()
```

1) Considere nuestro grafo de CR (provincias). Use el algoritmo 'greedy_color' de networkx para calcular la coloración.

```python
import networkx as nx

A = 'Alajuela'
H = 'Heredia'
SJ = 'San Jose'
G = 'Guanacaste'
P = 'Puntarenas'
C = 'Cartago'
L = 'Limon'

V = [A, H, SJ, C, P, L, G]
E = [(G,P), (G,A), (A,H), (A, SJ), (A, P), (H, SJ), (H, L), (SJ, P), (SJ, C), (SJ, L), (P, L), (C, L)]

costaRicaG = nx.Graph(E)
nx.draw(costaRicaG, with_labels=True)
plt.show()


costaRicaGColored = nx.coloring.greedy_color(costaRicaG, strategy="largest_first")

print(costaRicaGColored)
```
